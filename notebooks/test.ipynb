```python
import pandas as pd

path_A = r"Y:\Python\Laptop-Price-Prediction\data\raw\chotot_laptop_data.csv"
path_B = r"Y:\Python\Laptop-Price-Prediction\data\raw\clean_laptop_features.csv"

df_A = pd.read_csv(path_A)
df_B = pd.read_csv(path_B)
```

```python
print("=== Dataset A ===")
print("Shape:", df_A.shape)
print("Columns:", list(df_A.columns))

print("\n=== Dataset B ===")
print("Shape:", df_B.shape)
print("Columns:", list(df_B.columns))
```

```python
def quick_info(df, name):
    print(f"\n=== {name} INFO ===")
    print(df.dtypes)
    print("\nMissing values:")
    print(df.isnull().sum().sort_values(ascending=False).head(10))

quick_info(df_A, "Dataset A")
quick_info(df_B, "Dataset B")
```

```python
print("=== Sample A ===")
display(df_A.head(3))

print("=== Sample B ===")
display(df_B.head(3))
```

```python
cols_A = set(df_A.columns)
cols_B = set(df_B.columns)

common_cols = cols_A & cols_B
only_A = cols_A - cols_B
only_B = cols_B - cols_A

print("Common columns:", common_cols)
print("\nOnly in A:", only_A)
print("\nOnly in B:", only_B)
```

```python
for col in common_cols:
    if df_A[col].dtype == 'object' or df_B[col].dtype == 'object':
        print(f"\n=== Column: {col} ===")
        print("A sample:", df_A[col].dropna().astype(str).head(3).tolist())
        print("B sample:", df_B[col].dropna().astype(str).head(3).tolist())
```

```python
print("=== Numeric summary A ===")
display(df_A.describe())

print("=== Numeric summary B ===")
display(df_B.describe())
```
